## Sigmoid Function



```python
import math

def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```

```python
print('sigmoid(100) : ', sigmoid(100))
```

## Tanh Function

```python
def tanh(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
```

```python
print('tanh(100) : ', tanh(100))
print('tanh(50) : ', tanh(50))
print('tanh(-100) : ', tanh(-100))
```

## Relu Function

```python
def relu(x):
    return max(0,x)
```

```python
print('relu(100) : ', relu(100))
print('relu(50) : ', relu(50))
print('relu(-100) : ', relu(-100))
```

## LeakyRelu Function

```python
def leaky_relu(x):
    return max(0.1*x, x)
```

```python
print('leaky_relu(100) : ', leaky_relu(100))
print('leaky_relu(50) : ', leaky_relu(50))
print('leaky_relu(-100) : ', leaky_relu(-100))
```
