<h3 style='color:blue'> Calculate profit and loss</h3>

```python
revenue = np.array([[180,200,220],[24,36,40],[12,18,20]])
expenses = np.array([[80,90,100],[10,16,20],[8,10,10]])
```

```python
profit = revenue - expenses
profit
```

<h3 style='color:blue'>Calculate total sales from units and price per unit using matrix multiplication</h3>

```python
price_per_unit = np.array([1000,400,1200])
units = np.array([[30,40,50],[5,10,15],[2,5,7]])
```

<p style='color:gray'>
This is done as like <br>
price_per_unit = np.array([[1000,400,1200], [1000,400,1200], [1000,400,1200]])<br>
units = np.array([[30,40,50],[5,10,15],[2,5,7]]) <br>
price_per_unit*units</p>

```python
price_per_unit*units
```

```python
total_sales = np.dot(price_per_unit,units)
total_sales
```
